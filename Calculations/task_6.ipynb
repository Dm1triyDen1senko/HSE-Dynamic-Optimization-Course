```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
# Задание 6

# Для начала инициализируем произвольные начальные условия:

start_position = 8  # начальная позиция системы
start_velocity = 64  # начальная скорость системы

simulation_duration = 1.0  # общее время прохождения симуляции
step_size = 0.01  # шаг дискретизации по времени
initial_guess = 0.0  # начальное значение управления
bounds = [(-1, 1)]  # ограничения на управление


# Далее определим функцию, вычисляющущю эффективность управления:

def control_efficiency(control):
    time_points = np.arange(0, simulation_duration, step_size)  # создадим "временную сетку"
    positions = np.zeros_like(time_points)  # создадим массив для хранения позиций
    velocities = np.zeros_like(time_points)  # также создадим массив для хранения скоростей

    # Установим введенные раннее начальные условия:
    positions[0] = start_position
    velocities[0] = start_velocity

    # Реализуем итеративное обновление состояния системы на основе управления:
    for idx in range(1, len(time_points)):
        velocities[idx] = velocities[idx - 1] + control * step_size  # здесь происходит обновление скорости
        positions[idx] = positions[idx - 1] + velocities[idx - 1] * step_size  # здесь - обновление позиции

    # Вычисление ошибки в конечный момент времени:
    terminal_error = np.abs(positions[-1]) + np.abs(velocities[-1])
    return terminal_error

# Произведем оптимизацию для нахождения наилучшего управления:

optimization_result = minimize(control_efficiency, initial_guess, bounds=bounds)

# Извлечем значения для потимального управления и времени:

best_control = optimization_result.x[0]  
best_time = optimization_result.fun

# Проведем повторную симуляцию с использованием найденного значения для наилучшего управления:

time_points = np.arange(0, simulation_duration, step_size)
positions = np.zeros_like(time_points)  
velocities = np.zeros_like(time_points) 

positions[0] = start_position
velocities[0] = start_velocity

for idx in range(1, len(time_points)):
    velocities[idx] = velocities[idx - 1] + best_control * step_size  
    positions[idx] = positions[idx - 1] + velocities[idx - 1] * step_size
    
# Ну и наконец выведем результаты:
    
print(f'Результаты:')
print()
print(f'Наилучшее управление: {round(best_control, 2)}')
print(f'Время T: {round(best_time, 2)}')

plt.figure(figsize=(10, 6))
plt.plot(time_points, positions, label='Позиция', color='blue')  
plt.xlabel('Время')  
plt.ylabel('Позиция') 
plt.title('Изменение позиции')  
plt.grid(True)  
plt.legend()  
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(time_points, velocities, label='Скорость', color='green')
plt.xlabel('Время')  
plt.ylabel('Скорость')  
plt.title('Изменение скорости') 
plt.grid(True)  
plt.legend()  
plt.show()
```
